# Data Exploration

- Here, we just illustrate that Data Scientists would connect via pd.read_sql() to the DWH.
- This allows them to visualize data, run ML models, explore correlations, etc.

```python
import pandas as pd
import seaborn as sns
from sqlalchemy import create_engine
from local_env import CONN_STR


engine = create_engine(CONN_STR)
df = pd.read_sql("select * from dbt.users", con=engine)

print(df.columns)

sns.regplot(data=df, x="unpaid_subscriptions_cnt", y="lifetime_revenue").set(title="Useless Graph Just for Illustration Purposes")
```
